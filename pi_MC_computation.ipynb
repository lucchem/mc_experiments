```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Computing $\pi$ via Monte Carlo

The approximation of $\pi$ is a classical Monte Carlo example. Draw random points in $(0,1)\times(0,1)$, and consider that the ratio of points within one radius from the origin is the $\pi/4$ of the total.

```python
exp_max   = 7

n         = 10**exp_max
points    = np.random.uniform(0,1,(2,n))
distances = points[0]**2+points[1]**2 < 1

problem_sizes = np.array([10**j for j in range(2,exp_max+1)])

plot = []
for i in problem_sizes:
    circle = sum(distances[:i])
    plot.append(4*circle/i)
```

The approximation can be compared to the exact value by means of L$_1$-error, and one more or less gets the expected Monte Carlo-CLT scaling ($y\propto \sqrt x$).

```python
plt.xscale('log')
plt.yscale('log')
plt.xlabel('sample size')
plt.ylabel(r'L$_1$-error')
plt.xlim((40,1.8*n))
plt.ylim((10e-5,1))
plt.scatter(problem_sizes,np.abs(np.array(plot)-math.pi),label='L1 error')
plt.plot(problem_sizes,problem_sizes**-0.5,label=r'$y\propto \sqrt{{x}}$')
plt.legend()
plt.show()
```
