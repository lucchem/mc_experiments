```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
dt = 1/N
B = np.zeros(N+1)
C = np.zeros(N+1)
t = np.zeros(N+1)
W = np.random.normal(loc = 0, scale = 1, size = N)
k=1
h=5

for i in range(1,N+1):
    t[i] = i*dt
    B[i]=B[i-1] -   B[i-1]/(1-t[i-1])*dt +   np.sqrt(dt)*W[i-1]
    C[i]=C[i-1] - h*C[i-1]/(1-t[i-1])*dt + k*np.sqrt(dt)*W[i-1]

plt.plot(t,B,label='normal')
plt.plot(t,C,label='rescaled')
plt.legend()
plt.show()
```
