```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
```

# Verifying the properties of the Box-Muller transform

Refer to wikipedia for a quick explanation of the method. It consists in getting two normal variables ``x`` and ``y`` from two uniform variables ``v`` and ``w``.

Here, I just get 1 million samples and verify the distribution of ``x``.

```python
v = np.random.uniform(0,1,1000000)
w = np.random.uniform(0,1,1000000)
x = np.sqrt(-2*np.log(v))*np.cos(2*math.pi*w)
```

```python
count, bins, ignored = plt.hist(x, 100, density=True, label='Box-Mueller transform')
plt.plot(bins, sts.norm.pdf(bins), linewidth=1, color='r',label='Gaussian')
plt.legend()
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()
```
